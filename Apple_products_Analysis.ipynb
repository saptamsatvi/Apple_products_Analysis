```python
#TOP -5 PRODUCTS...........
```

```python
top_products_by_price = df.nlargest(5, 'Sale Price')
print("Top 5 Products by Sale Price:\n", top_products_by_price[['Product Name', 'Sale Price']])
```

```python
top_products_by_price = df.nlargest(5, 'Sale Price')

plt.figure(figsize=(5, 3))
sns.barplot(x='Sale Price', y='Product Name', data=top_products_by_price)
plt.title('Top 5 Products by Sale Price')
plt.xlabel('Sale Price (INR)')
plt.ylabel('Product Name')
plt.show()
```

```python
#AVERAGE SATR RATING 
```

```python
average_star_rating = df['Star Rating'].mean()
print("Average Star Rating:", average_star_rating)
```

```python
verage_star_rating = df['Star Rating'].mean()

plt.figure(figsize=(1, 5))
sns.barplot(x=['Average Star Rating'], y=[average_star_rating])
plt.title('Average Star Rating')
plt.ylim(0, 5) 
plt.ylabel('Star Rating')
plt.show()
```

```python
# Average Discount Percentage
```

```python
average_discount = df['Discount Percentage'].mean()
print("Average Discount Percentage:", average_discount)
```

```python
average_discount = df['Discount Percentage'].mean()

plt.figure(figsize=(2, 5))
sns.barplot(x=['Average Discount Percentage'], y=[average_discount])
plt.title('Average Discount Percentage')
plt.ylabel('Discount Percentage (%)')
plt.show()
```

```python
# Top Products by Number of Reviews

```

```python
top_products_by_reviews = df.nlargest(5, 'Number Of Reviews')
print("Top Products by Number of Reviews:\n", top_products_by_reviews[['Product Name', 'Number Of Reviews']])
```

```python
top_products_by_reviews = df.nlargest(5, 'Number Of Reviews')

plt.figure(figsize=(10, 3))
sns.barplot(x='Number Of Reviews', y='Product Name',hue='Product Name', data=top_products_by_reviews, palette='magma')
plt.title('Top Products by Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Product Name')
plt.show()
```
